```python
# Creating duplicate dataset for creating model to find feature importances

df1 = df.copy()
```

```python
# Binary encoding of columns having only two values

df1['attrition'] = df1['attrition'].apply(lambda x:1 if x=='Yes' else 0)
df1['gender'] = df1['gender'].apply(lambda x:1 if x=='Male' else 0)
df1['overtime'] = df1['overtime'].apply(lambda x:1 if x=='Yes' else 0)
```

```python
df1.head()
```

```python
# Label encoding the data having three or more values

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df1['btravel_label'] = le.fit_transform(df1['businesstravel'])
df1['department_label'] = le.fit_transform(df1['department'])
df1['educationfield_label'] = le.fit_transform(df1['educationfield'])
df1['jobrole_label'] = le.fit_transform(df1['jobrole'])
df1['maritalstatus_label'] = le.fit_transform(df1['maritalstatus'])
```

```python
df1 = df1.drop(columns=['businesstravel','department','educationfield','jobrole','maritalstatus'],axis = 1)
```

```python
df1.head()
```

```python
# Model Training

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

x,y = df1.drop(columns=['attrition'],axis = 1), df1['attrition']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)

model = RandomForestClassifier(n_jobs=-1)

model.fit(x_train, y_train)
```

```python
model.score(x_test,y_test)
```

```python
# Finding the feature importances

sorted_importances = dict(sorted(zip(model.feature_names_in_, model.feature_importances_), key = lambda x:x[1], reverse=True))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,8))
plt.bar(sorted_importances. keys(), sorted_importances.values())
plt.xticks(rotation=45, ha='right')

plt. show()
```
